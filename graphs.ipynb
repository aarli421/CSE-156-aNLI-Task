```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

models = df["model"].unique()

for model in models:
    model_data = df[df["model"] == model]
    plt.figure()
    plt.plot(model_data["iteration"], model_data["loss"], marker="o", linestyle="-")
    plt.xlabel("Iteration")
    plt.ylabel("Loss")
    plt.title(f"{model} - Iteration vs Loss")
    plt.grid(True)
    plt.show()
```
